# Transforming Data

```python
#splitting hero & publisher
hero_info[['Hero', 'Publisher']] = hero_info['Hero|Publisher'].str.split('|',expand = True)
hero_info.head()
```

```python
#dropping Hero|Publisher
hero_info = hero_info.drop(columns='Hero|Publisher')
```

```python
hero_info.head()
```

```python
#testing measurement value
measurements = hero_info.loc[0, 'Measurements']
print(type(measurements))
measurements
```

```python
#replacing '' with "" 
measurements = measurements.replace("'", '"')
measurements
```

```python
#making dict from string on test
fixed_measurements = json.loads(measurements)
print(type(fixed_measurements))
fixed_measurements
```

```python
#replacing '' with "" for column
hero_info['Measurements'] = hero_info['Measurements'].str.replace("'", '"')
```

```python
#making dict from string on column
hero_info['Measurements'] = hero_info['Measurements'].apply(json.loads)
hero_info['Measurements'].head()
```

```python
#using test sample
test_meas = hero_info.loc[0, 'Measurements']
print(type(test_meas))
test_meas
```

```python
#applying od.series to column
height_weight = hero_info['Measurements'].apply(pd.Series)
height_weight
```

```python
#concat height weight to main df
hero_info = pd.concat((hero_info, height_weight), axis=1)
hero_info.head(2)
```

```python
#dropping measurements column
hero_info = hero_info.drop(columns='Measurements')
```

```python
#splitting number from cm
hero_info[['Height_cm', 'cm']] = hero_info['Height'].str.split(' ', expand= True)
```

```python
#changing from string to float
hero_info['Height_cm'] = hero_info['Height_cm'].astype(float)
```

```python
#splitting number from kg for Weight
hero_info[['Weight_kg', 'kg']] = hero_info['Weight'].str.split(' ', expand= True)
```

```python
#changing from string to float for Weight
hero_info['Weight_kg'] = hero_info['Weight_kg'].astype(float)
```

```python
#dropping remaining columns not needed 
hero_info = hero_info.drop(columns= ['Height', 'cm', 'Weight', 'kg'])
```

```python
#reordering columns
hero_info = hero_info[['Hero', 'Publisher', 'Gender', 'Race', 'Alignment', 'Hair color', 
           'Eye color', 'Skin color', 'Height_cm', 'Weight_kg']]
```

```python
#checking columns
hero_info.head(2)
```

# Transforming Data Part 2

```python
hero_powers.head()
```

```python
#checking type
type(hero_powers['Powers'])
```

```python
#splitting powers
hero_powers['pow'] = hero_powers['Powers'].str.split(',', expand=False)
hero_powers['pow']
```

```python
#transforming each element of power list column to a row using explode
exploded = hero_powers.explode('pow')
exploded
```

```python
#finding the unique powers
pow_to_make = exploded['pow'].dropna().unique()
```

```python
#making function for unique pows to be in columns
for pow in pow_to_make:
    hero_powers[pow] = hero_powers['Powers'].str.contains(pow)
```

```python
#dropping power and pow columns
hero_powers = hero_powers.drop(columns= ['Powers', 'pow'])
```

```python
#checking columns
hero_powers.head()
```

# Combining DataFrames 

```python
#merging hero_info and hero_power using hero_name
df_heros = pd.merge(hero_info, hero_powers, left_on='Hero', right_on='hero_names')
```

```python
#checking df
df_heros.head()
```

```python
#checking df info
df_heros.info()
```

```python
#saving df to Data
df_heros.to_csv('Data/superhero_combined_final.csv.gz', index_label=False)
```
